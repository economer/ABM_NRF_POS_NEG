```python
%%R

df_results = df_results %>% 
mutate(nrf = pos - neg) %>% 
 mutate(nit_r_factor = case_when(
    nit_rate %in% c(0) ~ "Universal",
    nit_rate %in% c(0.3) ~ "NIT: 0.3",
    nit_rate %in% c(0.5) ~ "NIT: 0.5",
    nit_rate %in% c(0.8) ~ "NIT: 0.8",
   T ~ "Top-Up"
 )) 

```

```python
%%R -w 1200 -h 1000
# average nrf 
df_results %>% 
#filter(`Base_Benefit`==46432/26) %>%
mutate(nrf = pos - neg) %>%
group_by(`Step`,nit_r_factor) %>%
summarize(mean_nrf = mean(nrf,na.rm=T)) %>%
ggplot(aes(`Step`,mean_nrf)) +
geom_line(aes(color=nit_r_factor),size=1) +
geom_point() +
#facet_wrap(~nit_r_factor) +
theme_bw()
```

```python
%%R -w 1200 -h 1000
# median nrf 
df_results %>% 
#filter(`Base_Benefit`==46432/26) %>%
mutate(nrf = pos - neg) %>%
group_by(`Step`,nit_r_factor) %>%
summarize(mean_nrf = median(nrf,na.rm=T)) %>%
ggplot(aes(`Step`,mean_nrf)) +
geom_line(aes(color=nit_r_factor),size=1) +
geom_point() +
#facet_wrap(~nit_r_factor) +
theme_bw()
```

```python
%%R -w 1200 -h 1000
# average pos 
df_results %>% 
#filter(`Base_Benefit`==46432/26) %>%
mutate(nrf = pos - neg) %>%
group_by(`Step`,nit_r_factor) %>%
summarize(mean_pos = mean(pos,na.rm=T)) %>%
ggplot(aes(`Step`,mean_pos)) +
geom_line(aes(color=nit_r_factor),size=1) +
geom_point() +
#facet_wrap(~nit_r_factor) +
theme_bw()
```

```python
%%R -w 1200 -h 1000
# average neg 
df_results %>% 
#filter(`Base_Benefit`==46432/26) %>%
mutate(nrf = pos - neg) %>%
group_by(`Step`,nit_r_factor) %>%
summarize(mean_neg = mean(neg,na.rm=T)) %>%
ggplot(aes(`Step`,mean_neg)) +
geom_line(aes(color=nit_r_factor),size=1) +
geom_point() +
#facet_wrap(~nit_r_factor) +
theme_bw()
```
